```python
# Análise: Impacto das Horas Extras na Saída
print("\n--- Gerando Gráfico: Impacto das Horas Extras na Saída ---")

# Dicionário de Tradução
dicionario_saida = {'No': 'Ficou', 'Yes': 'Saiu'}
df_limpo['Situacao'] = df_limpo['Attrition'].map(dicionario_saida)

dicionario_overtime = {'No': 'Não', 'Yes': 'Sim'}
df_limpo['HorasExtras_PT'] = df_limpo['OverTime'].map(dicionario_overtime)


# Gerando o Gráfico com as Colunas Traduzidas
ax = sns.countplot(x='HorasExtras_PT', hue='Situacao', data=df_limpo, order=['Não', 'Sim'])

# Adicionando os rótulos em cada barra
for container in ax.containers:
    ax.bar_label(container)

# Configurando títulos e rótulos
plt.title('Saída de Funcionários vs. Realização de Horas Extras', weight='bold')
plt.ylabel('Número de Colaboradores', fontsize=12)
plt.xlabel('Faz Horas Extras?', fontsize=12)
plt.legend(title='Situação')

# Salvando o gráfico
caminho_graficos = r'C:\Users\Aline.Aline\Desktop\GitHub\Projeto 8 - Análise de Fatores IBM RH\graficos'
nome_do_arquivo = 'barras_overtime_vs_saida.png'
plt.savefig(f'{caminho_graficos}\\{nome_do_arquivo}', bbox_inches='tight', dpi=300)

plt.show()

print(f"Gráfico '{nome_do_arquivo}' salvo com sucesso!")
```

## ⏰ Insights do Custo das Horas Extras ⏰

***

### 🔥 O Risco Multiplicado por 3

* **O que isso significa?** Um funcionário que faz horas extras tem **praticamente 3 vezes mais chance** de pedir demissão. A sobrecarga de trabalho não é apenas um fator de desgaste, é um dos principais catalisadores do *turnover* nesta empresa. Este é um dos indicadores preditivos mais fortes que encontramos.

***

### 🎯 O Impacto Desproporcional na Retenção

* **Observação:** O grupo que faz horas extras representa apenas **28%** da força de trabalho total da empresa (416 de 1470 funcionários).
* **O que isso significa?** Embora sejam minoria na empresa, este pequeno grupo de 28% é responsável por **mais da metade (53%) de todas as saídas**. Combater a cultura de horas extras é a ação com o maior potencial de impacto para reduzir o *turnover* geral da empresa.

***

### 💸 O Custo Oculto da Sobrecarga

* **Observação:** O número absoluto de pessoas que fazem horas extras e saem (127) é maior do que o número de pessoas que não fazem e saem (110).
* **O que isso significa?** A empresa está, literalmente, perdendo mais gente por esgotamento e burnout do que por uma "insatisfação geral" durante o horário normal de trabalho. Isso reforça que as saídas não são apenas sobre salário ou cargo, mas sobre **bem-estar e a sustentabilidade do ritmo de trabalho**. É um problema direto de saúde organizacional.

```python
# Análise: Correlação de cada fator com a Saída
print("\n--- Gráfico de Correlação Direta com a Saída ---")

# Calculando a correlação
df_numerico = df_limpo.select_dtypes(include=['int64', 'float64'])
df_numerico['Situacao_Num'] = df_limpo['Situacao'].apply(lambda x: 1 if x == 'Saiu' else 0)
matriz_correlacao = df_numerico.corr()
corr_saida = matriz_correlacao['Situacao_Num'].sort_values(ascending=False)

# Traduzindo os Fatores (usando o mesmo dicionário que foi criado para os histogramas)
dicionario_titulos = {
    'Situacao_Num': 'Situação (Numérico)', 'Age': 'Idade', 'DailyRate': 'Taxa Diária',
    'DistanceFromHome': 'Distância de Casa', 'Education': 'Nível de Educação',
    'EnvironmentSatisfaction': 'Satisfação com Ambiente', 'HourlyRate': 'Taxa Horária',
    'JobInvolvement': 'Envolvimento no Trabalho', 'JobLevel': 'Nível do Cargo',
    'JobSatisfaction': 'Satisfação no Trabalho', 'MonthlyIncome': 'Salário Mensal',
    'MonthlyRate': 'Taxa Mensal', 'NumCompaniesWorked': 'Nº de Empresas Anteriores',
    'PercentSalaryHike': '% de Aumento Salarial', 'PerformanceRating': 'Nota de Performance',
    'RelationshipSatisfaction': 'Satisfação nos Relacionamentos', 'StockOptionLevel': 'Nível de Ações',
    'TotalWorkingYears': 'Total de Anos de Carreira', 'TrainingTimesLastYear': 'Treinamentos no Último Ano',
    'WorkLifeBalance': 'Equilíbrio Vida-Trabalho', 'YearsAtCompany': 'Anos na Empresa',
    'YearsInCurrentRole': 'Anos no Cargo Atual', 'YearsSinceLastPromotion': 'Anos desde a Última Promoção',
    'YearsWithCurrManager': 'Anos com o Gestor Atual'
}

# Renomeando o índice da série de correlações
corr_saida.index = corr_saida.index.map(dicionario_titulos)


# Gerando o gráfico com os nomes já traduzidos
plt.figure(figsize=(10, 10)) # Aumentei a altura para caber todos os nomes
ax = sns.barplot(x=corr_saida.values, y=corr_saida.index)

plt.title('Força da Correlação de Cada Fator com a Saída do Colaborador', weight='bold')
plt.xlabel('Coeficiente de Correlação', fontsize=12)
plt.ylabel('Fatores Analisados', fontsize=12)

# Adicionando os valores nas barras para clareza
ax.bar_label(ax.containers[0], fmt='%.2f', padding=3)

# Salvando o gráfico
caminho_graficos = r'C:\Users\Aline.Aline\Desktop\GitHub\Projeto 8 - Análise de Fatores IBM RH\graficos'
nome_do_arquivo = 'barras_correlacao_saida.png'
plt.savefig(f'{caminho_graficos}\\{nome_do_arquivo}', bbox_inches='tight', dpi=300)

plt.show()

print(f"Gráfico '{nome_do_arquivo}' salvo com sucesso!")
```

Análise final que resume a força e a direção da relação de cada fator numérico com a saída de um colaborador.

***

### 🏛️ Os Pilares da Retenção: Senioridade e Salário

* **Observação:** As correlações negativas mais fortes (os maiores valores negativos) são com **Total de Anos de Carreira (-0.17)**, **Nível do Cargo (-0.17)**, **Salário Mensal (-0.16)** e **Idade (-0.16)**.
* **O que isso significa?** Este é o perfil claro do funcionário que **fica**: mais velho, com mais experiência, em um cargo mais alto e com maior salário. A falta de um desses pilares, especialmente o salário e a progressão de cargo, é o principal fator de risco para a saída. O gráfico confirma tudo que vimos nos boxplots e histogramas.

### 🤝 O Poder do Vínculo e da Estabilidade

* **Observação:** **Anos na Empresa (-0.13)**, **Anos no Cargo Atual (-0.16)** e **Anos com o Gestor Atual (-0.16)** também mostram uma correlação negativa relevante.
* **O que isso significa?** O tempo cria um vínculo que ajuda a reter os funcionários. A rotatividade é maior no início do "ciclo" de um colaborador na empresa, no cargo ou com um novo gestor. * **Observação:** Fatores como **% de Aumento Salarial (-0.01)**, **Equilíbrio Vida-Trabalho (-0.06)** e **Distância de Casa (0.08)** possuem uma correlação muito próxima de zero.

    * Não adianta dar pequenos aumentos se o **salário base** for baixo; o problema é estrutural.
    * Nesta empresa, o equilíbrio vida-trabalho e a distância de casa, embora importantes, **não são os principais motivos** que levam à demissão, ao contrário do que o senso comum poderia sugerir. O foco da empresa deve estar nos problemas mais impactantes de carreira e remuneração.
